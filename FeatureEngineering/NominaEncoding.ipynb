```python
# Frequency encoding
```

```python
def frequency_encoding(df, feature, top_no_feature) :
    freqency = df[feature].value_counts().sort_values(ascending = False).index.tolist()[:top_no_feature]
    for i in freqency :
        df[feature + "_"  + i] = np.where(df[feature] == i, 1, 0)
    print("[+]Task completed..")
```

```python
frequency_encoding(df, 'X0', 3)
```

```python
# result of frequency encoding
df
```

```python
# Count Encoding
```

```python
value_index = {
    
}
def count_encoding(df, feature) :
    count = df[feature].value_counts()
    indexes = count.index
    values = [word for word in count]
    for i in range(len(values)) :
        value_index[indexes[i]] = values[i]
    df[feature] = df[feature].map(value_index)
    print("[+] Task Completed..")
```

```python
count_encoding(df, 'X0')
```

```python
# result of count encoding
df
```

```python
# Mean Encoding
```

```python
value_index = {
    
}
def mean_encoding(df, feature) :
    count = df[feature].value_counts()
    index = count.index
    mean = count / len(df)
    mean = [m for m in mean]
    for i in range(len(count)) :
        value_index[index[i]] = mean[i]
    df[feature] = df[feature].map(value_index)
    print("[+] Completed...")
```

```python
mean_encoding(df, 'X1')
```
